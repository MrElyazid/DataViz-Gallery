```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('iris.data', delimiter=',', names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])

pd.set_option("display.max_rows", 50)
df
```

```python
setosa_df = df[df['class'] == 'Iris-setosa']
versicolor_df = df[df['class'] == 'Iris-versicolor']
virginica_df = df[df['class'] == 'Iris-virginica']
```

# Plotting the features for each class ( Setosa, versicolor, virginica )

## Setosa : 

```python


fig, axs = plt.subplots(2, 2, figsize=(16, 10))
step = 0.1

# Sepal length
xmin00, xmax00 = setosa_df['sepal_length'].min(), setosa_df['sepal_length'].max()
axs[0, 0].hist(setosa_df['sepal_length'], bins=np.arange(xmin00, xmax00 + 2*step, step), linewidth=0.25, edgecolor="white")
axs[0, 0].set_xticks(np.arange(xmin00, xmax00 + step, step))
axs[0, 0].set_title("Sepal Length Distribution", fontsize=16)
axs[0, 0].set_xlabel("Sepal Length (cm)", fontsize=14)
axs[0, 0].set_ylabel("Count", fontsize=14)
axs[0, 0].grid(True, linestyle='--', alpha=0.7)

# Sepal width
xmin01, xmax01 = setosa_df['sepal_width'].min(), setosa_df['sepal_width'].max()
axs[0, 1].hist(setosa_df['sepal_width'], bins=np.arange(xmin01, xmax01 + 2*step, step), linewidth=0.25, edgecolor="white")
axs[0, 1].set_xticks(np.arange(xmin01, xmax01 + step, step))
axs[0, 1].set_title("Sepal Width Distribution", fontsize=16)
axs[0, 1].set_xlabel("Sepal Width (cm)", fontsize=14)
axs[0, 1].set_ylabel("Count", fontsize=14)
axs[0, 1].grid(True, linestyle='--', alpha=0.7)

# Petal length
xmin10, xmax10 = setosa_df['petal_length'].min(), setosa_df['petal_length'].max()
axs[1, 0].hist(setosa_df['petal_length'], bins=np.arange(xmin10, xmax10 + 2*step, step), linewidth=0.25, edgecolor="white")
axs[1, 0].set_xticks(np.arange(xmin10, xmax10 + step, step))
axs[1, 0].set_title("Petal Length Distribution", fontsize=16)
axs[1, 0].set_xlabel("Petal Length (cm)", fontsize=14)
axs[1, 0].set_ylabel("Count", fontsize=14)
axs[1, 0].grid(True, linestyle='--', alpha=0.7)

# Petal width
xmin11, xmax11 = setosa_df['petal_width'].min(), setosa_df['petal_width'].max()
axs[1, 1].hist(setosa_df['petal_width'], bins=np.arange(xmin11, xmax11 + 2*step, step), linewidth=0.25, edgecolor="white")
axs[1, 1].set_xticks(np.arange(xmin11, xmax11 + step, step))
axs[1, 1].set_title("Petal Width Distribution", fontsize=16)
axs[1, 1].set_xlabel("Petal Width (cm)", fontsize=14)
axs[1, 1].set_ylabel("Count", fontsize=14)
axs[1, 1].grid(True, linestyle='--', alpha=0.7)

fig.suptitle('Distribution of Sepal and Petal Dimensions for Setosa Class', fontsize=22)
plt.tight_layout(rect=[0, 0, 1, 0.96])
```

# calculating averages in each class : 

```python



petal_width = sum(df['petal_width'])/len(df['petal_width'])
petal_length = sum(df['petal_length'])/len(df['petal_length'])
sepal_width = sum(df['sepal_width'])/len(df['sepal_width'])
sepal_length = sum(df['sepal_length'])/len(df['sepal_length'])


setosa_petal_width = sum(setosa_df['petal_width'])/len(setosa_df['petal_width'])
setosa_petal_length = sum(setosa_df['petal_length'])/len(setosa_df['petal_length'])
setosa_sepal_width = sum(setosa_df['sepal_width'])/len(setosa_df['sepal_width'])
setosa_sepal_length = sum(setosa_df['sepal_length'])/len(setosa_df['sepal_length'])


versicolor_petal_width = sum(versicolor_df['petal_width'])/len(versicolor_df['petal_width'])
versicolor_petal_length = sum(versicolor_df['petal_length'])/len(versicolor_df['petal_length'])
versicolor_sepal_width = sum(versicolor_df['sepal_width'])/len(versicolor_df['sepal_width'])
versicolor_sepal_length = sum(versicolor_df['sepal_length'])/len(versicolor_df['sepal_length'])

virginica_petal_width = sum(virginica_df['petal_width'])/len(virginica_df['petal_width'])
virginica_petal_length = sum(virginica_df['petal_length'])/len(virginica_df['petal_length'])
virginica_sepal_width = sum(virginica_df['sepal_width'])/len(virginica_df['sepal_width'])
virginica_sepal_length = sum(virginica_df['sepal_length'])/len(virginica_df['sepal_length'])
```

# grouped bar chart to show differences in averages across categories :

```python

categories = ['Petal Width', 'Petal Length', 'Sepal Width', 'Sepal Length']

data = {
    'overall_avg': [petal_width, petal_length, sepal_width, sepal_length],
    'setosa_avg': [setosa_petal_width, setosa_petal_length, setosa_sepal_width, setosa_sepal_length],
    'versicolor_avg': [versicolor_petal_width, versicolor_petal_length, versicolor_sepal_width, versicolor_sepal_length],
    'virginica_avg': [virginica_petal_width, virginica_petal_length, virginica_sepal_width, virginica_sepal_length]
}

x = np.arange(len(categories))  
width = 0.25  
spacing = 0.05  
multiplier = 0


x_with_spacing = x * (1 + width + spacing)

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in data.items():
    offset = width * multiplier
    rects = ax.bar(x_with_spacing + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

ax.set_ylabel('Length (cm)')
ax.set_title('comparaison of averages across categories')
ax.set_xticks(x_with_spacing + width + 2*spacing, categories)
ax.legend(loc='upper left')


plt.show()
```

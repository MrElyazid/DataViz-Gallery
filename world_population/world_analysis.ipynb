Dataset link : https://www.kaggle.com/datasets/iamsouravbanerjee/world-population-dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

Dataset exploration : 

```python
df = pd.read_csv('world_population.csv')
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
```

Top 10 countries by population for 2022 : 

```python
top_10 = df.nlargest(10, '2022 Population')

plt.figure(figsize=(10, 6))
plt.bar(top_10['Country/Territory'], top_10['2022 Population'], alpha=0.7, color='blue', edgecolor='black')
plt.title('Top 10 Countries by Population (2022)')
plt.xlabel('Country')
plt.ylabel('Population (2022)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

Top 10 countries by population influence on the total human population in the world ( Pie chart ) :

```python

palette = sns.color_palette("husl", 11)



top_10_total = top_10['2022 Population'].sum()

other_total = df['2022 Population'].sum() - top_10_total

pie_data = pd.concat([top_10[['Country/Territory', '2022 Population']], 
                      pd.DataFrame({'Country/Territory': ['Other'], '2022 Population': [other_total]})])

plt.figure(figsize=(10, 8))  
plt.pie(pie_data['2022 Population'], labels=pie_data['Country/Territory'], autopct='%1.1f%%', colors=palette, 
        wedgeprops={'width': 0.3},  
        textprops={'family': 'sans-serif', 'fontweight': 'bold'})
plt.title('Distribution of 2022 Population by Country')
plt.tight_layout()
plt.axis('equal')
plt.show()
```
